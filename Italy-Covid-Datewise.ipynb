# What is the total number of reported cases and death related to Covid 19 in italy

```python
total_cases = covid_df.new_cases.sum()
total_deaths = covid_df.new_deaths.sum()
```

```python
print('The number of reported cases is {} and the number of reported deaths {}'.format(total_cases,total_deaths))
```

# What is the overall death reate (ratio of reported deaths to reported cases)?

```python
death_rate = covid_df.new_deaths.sum() / covid_df.new_cases.sum()
```

# print('The overall reported death rate in italy is  {:.2f} %'.format(death_rate))

# What is the number of overall tests conducted? A total of 935310 test were conducted before daily test number were being reported

```python
initial_tests = 935310
total_tests = initial_tests + covid_df.new_tests.sum()
```

```python
total_tests
```

# What fraction of test returned a positive result ?

```python
positive_rate = total_cases / total_tests
```

```python
print('{:.2f}% of tests in Italy lead toa positive diagnosis.'.format(positive_rate))
```

```python
#Merging Data 
```

```python
urlretrieve('https://gist.githubusercontent.com/aakashns/8684589ef4f266116cdce023377fc9c8/raw/99ce3826b2a9d1e6d0bde7e9e559fc8b6e9ac88b/locations.csv', 
            'locations.csv')
```

```python
locations_df = pd.read_csv('locations.csv')
```

```python
locations_df
```

```python
locations_df[locations_df.location == 'Italy']
```

```python
covid_df['location'] = 'Italy'
```

```python
covid_df
```

```python
merged_df = covid_df.merge(locations_df, on = 'location')
```

```python
merged_df
```

```python
merged_df['cases_per_million'] = merged_df.total_cases * 1e6 / merged_df.population
```

```python
merged_df['deaths_per_million'] = merged_df.total_deaths * 1e6 / merged_df.population
```

```python
merged_df['tests_per_million'] = merged_df.total_tests * 1e6 / merged_df.population
```

```python
merged_df
```

```python
result_df = merged_df[[
    'date',
    'new_cases',
    'total_cases',
    'new_deaths',
    'total_deaths',
    'new_tests',
    'total_tests',
    'cases_per_million',
    'deaths_per_million',
    'tests_per_million'
]]
```
